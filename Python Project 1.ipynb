# Exploratory Data Analysis

## Gender

```python
csv.columns
```

```python
sns.countplot(x="Gender",data=csv)
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = csv)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sales_gen = csv.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x="Gender" , y="Amount", data=sales_gen)
```

## Age

```python
ax = sns.countplot(data = csv, x = 'Age Group', hue="Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = csv.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = sales_age)
```

## State

```python
# total number of orders from top 10 states

sales_state = csv.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})   #Bars ki height and width set akrne ke liye
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales from top 10 states

sales_state = csv.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

## Marital Status

```python
ax = sns.countplot(data = csv, x = 'Marital_Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = csv.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```
